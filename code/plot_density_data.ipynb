# Import Packages

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
```

# Read Data

```python
def read_density_csv(level_num, area_num):
    return pd.read_csv(
        "../data/density_csvs/level_{}__area_{}.csv".format(level_num, area_num),
        names= ["Minute", "Crowd_density"]
    )
```

```python
df10 = read_density_csv(5, 10)
df30 = read_density_csv(5, 30)
df35 = read_density_csv(5, 35)
df36 = read_density_csv(5, 36)
df52 = read_density_csv(5, 52)
```

# Analyse Data

```python
def analyse_df(df):
    print(df.head())
    print(df.index)
    return plt.plot(df.Minute, df.Crowd_density)
```

```python
analyse_df(df10)
```

```python
analyse_df(df30)
```

# Analyse Together

### Add Camera Number To Each DF

```python
df10["Camera_num"] = 10
df30["Camera_num"] = 30
df35["Camera_num"] = 35
df36["Camera_num"] = 36
df52["Camera_num"] = 52
```

### Join All DFs Into One

```python
all_dfs = [df10, df30, df35, df36, df52]
df = pd.concat(all_dfs, axis=0, ignore_index=True)
df
```

### Plot Graph Of All Together

```python
colours = ["red", "blue", "green", "orange", "purple"]
for (name, group), colour in zip(df.groupby("Camera_num"), colours):
    plt.plot(group["Minute"], group["Crowd_density"], color=colour, label=name)

plt.legend(loc='best')
plt.show()
```
